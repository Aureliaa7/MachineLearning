```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data_path = '/kaggle/input/house-prices-advanced-regression-techniques/train.csv'
test_data_path = '/kaggle/input/house-prices-advanced-regression-techniques/test.csv'

train_data = pd.read_csv(train_data_path)
test_data = pd.read_csv(test_data_path)

print('Columns in the training set: ', len(train_data.columns))    
#print(train_data.columns)

print('\nColumns in the test set: ', len(test_data.columns))
#print(test_data.columns)
```

```python
#information about the number of rows, columns, column data types, memory usage
print(train_data.info())
#The columns with object dtype are the possible categorical features in dataset.
```

```python
# shape for train data and test data before one hot encoding
print(train_data.shape)
print(train_data)
```

```python
# one hot encoding for training data
train_data = pd.get_dummies(train_data)
print(train_data.shape)
print(train_data)
i = 0
for col in train_data.columns:
    print(col,  i)
    i += 1
# SalePrice is at index 37(counting from 0)
```

```python
# one hot encoding for test data
test_data = pd.get_dummies(test_data)
print(test_data.shape)

```

```python
# Handle missing values
#Imputation fills in the missing value with some number.
#The imputed value won't be exactly right in most cases, but it usually gives more accurate models than dropping the column entirely.


from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()

# handle missing values for training data
train_data = pd.DataFrame(my_imputer.fit_transform(train_data))
print(train_data.shape)
print(train_data)
```

```python
print(train_data.columns)
```

```python
# handle missing values for test data
test_data = pd.DataFrame(my_imputer.fit_transform(test_data))
print(test_data.shape)
print(test_data)
```

```python
# split the training data: I'll use a part for validating the model
```

```python
from sklearn.model_selection import train_test_split

features = test_data.columns
X = train_data[features]
y = train_data[37]
# test_size=the proportion of the dataset to include in the test split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.1, random_state=1)

print('Initial size of the dataset:', len(X))
print('The size of the dataset used for training the model: ',len(train_X))
print('The size of the dataset used for validating the model: ',len(test_X))
```

```python
### DECISION TREE REGRESSOR
```

```python
#fit model
from sklearn.tree import DecisionTreeRegressor

train_model = DecisionTreeRegressor(random_state = 1)
train_model.fit(train_X, train_y)

# predictions
predictions_dec_tree_regressor = train_model.predict(test_X)
#print('Predictions using DecisionTreeRegressor', predictions_dec_tree_regressor)
# model validation
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

print('Mean Absolute Error:', mean_absolute_error(test_y, predictions_dec_tree_regressor))
print('Mean Squared Error:', mean_squared_error(test_y, predictions_dec_tree_regressor))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(test_y, predictions_dec_tree_regressor)))
print('r2_score: ', r2_score(test_y, predictions_dec_tree_regressor))

```

```python
# LINEAR REGRESSION
# Scikit learn is used to import the linear regression model.
# The model is fit on the training data and it predicts the values for the testing data. 
# R2 score is used to measure the accuracy of the model.
```

```python
from sklearn.linear_model import LinearRegression 

linear_regressor = LinearRegression(normalize=True)
linear_regressor.fit(train_X, train_y)
pred_y = linear_regressor.predict(test_X)
print(r2_score(test_y,pred_y))
```

```python
# Evaluating the Algorithm
print('Mean Absolute Error:', mean_absolute_error(test_y, pred_y))
print('Mean Squared Error:', mean_squared_error(test_y, pred_y))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(test_y, pred_y)))
```
